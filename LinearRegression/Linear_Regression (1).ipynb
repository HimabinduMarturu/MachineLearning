```python
# Categorical Columns
a=df.select_dtypes(include='object').columns
for col in a:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, hue=None)
    plt.title(f'Distributio of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Numerical Columns
num_sta=df.select_dtypes(include=['int','float64']).columns
df[num_sta].hist(figsize=(30,20),bins=20,color='green',edgecolor='blue')
plt.xlabel('Value', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
plt.grid(True, linestyle='--', alpha=0.6)
plt.suptitle("Distribution of Numeric Variables", fontsize=16)
plt.show()
```
